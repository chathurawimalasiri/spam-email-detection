```python
# import libraries

import numpy as np
import pandas as pd 
```

```python
#Open the file
#This CSV file was taken from "https://www.kaggle.com/."

df = pd.read_csv('spam.csv', encoding = "ISO-8859-1")
df.head()
```

```python
# check the shape
# df.shape
df.columns
```

```python
# remove unwanted colmns

newdf = df.drop(["Unnamed: 2", 'Unnamed: 3', 'Unnamed: 4'],  axis='columns')
newdf.head()
```

```python
# check the null values

newdf.isna().sum()
```

```python
# add new colmn as "category". if spam = 1 and not spam = 0

newdf['category'] = newdf['v1'].apply(lambda x:1 if x=='spam' else 0)
newdf.head()
```

```python
# I divide the data set into two parts as train data and test data.

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(newdf.v2,newdf.category,test_size=0.3)
y_test
```

```python
newdf.info()
```

```python
# MultinomialNB implements the naive Bayes algorithm for multinomially distributed data, and is one of the 
# two classic naive Bayes variants used in text classification. 


from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

model = Pipeline([('vectorizer',CountVectorizer()),('nb',MultinomialNB())])
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(["Hi welcome!"])
```

```python
# create a model using joblib

import joblib

joblib.dump(model, 'email_spam_checker')
```

```python
# test the save model

SC = joblib.load('email_spam_checker')

SC.predict(['Congratulations! Youve been selected to receive a free cruise to the Bahamas! This is a limited time offer, so act now to claim your prize. To claim your prize, simply click on the link below and enter your personal information.'])
```
